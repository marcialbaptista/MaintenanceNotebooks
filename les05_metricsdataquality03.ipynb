### AE4465 (MM&A) - Lecture5 (Prognosability)

In prognostics and health management (PHM), it is imperative to have quality data. In PHM we have three popular methods to do monotonicity, trendability and prognosability calculation.
Prognosability is calculated as:

$$ prog =\exp( \frac{- std(x_j(N_j))} {mean\Big|x_j(1) - x_j(N_j)\Big|} ) $$

Where:
$x_j$ is a trajectory of measurements (vector)
$j$ is the jth system,
$N_j$ is the last measurement index of a trajectory

In this tutorial we will explore C-MAPSS case study and how to calculate the prognosability of a feature of signals.


```python
#import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import scipy.stats
```

### Data analysis

We can now open the CMAPSS dataset001 to perform some operations on the data.


```python
# Import the data in csv file
df = pd.read_csv('data/les05_CMAPSStrain001.txt', sep=' ')
# Read the columns
print('The columns of the csv file are: ')
df.head()
```

```python
# Let's start by plotting our time series.
for id_equipment in range(15):
    mask_equip1 = df['Equipment'] == id_equipment# Select column Equipment with value x
    signal = df.loc[mask_equip1,'2']
    plt.plot(range(len(signal[-50:])), signal[-50:] ) # Select feature 2 of unit 1
plt.show()
```

```python
# The sign or signum function is a standard mathematical function defined as sign(z) = z/|z|
def calculate_prognosability(feature_id):
    end_points = []
    ranges = []

    for i in range(1, 15, 1):
        mask_equip1 = df['Equipment'] == i
        signal = df.loc[mask_equip1, feature_id].values

        rnge = math.fabs(signal[-1] - signal[0])
        ranges.append(rnge)
        end_point = math.fabs(signal[-1])
        end_points.append(end_point)

    prognos = math.exp(-np.std(end_points) / np.mean(ranges))
    return prognos
```

### Function of prognosability

After defining the function of prognosability we can use it to calculate the prognosability of a feature.

```python
prog_1 = round(calculate_prognosability(feature_id='2'),3)
prog_2 = round(calculate_prognosability(feature_id='3'),3)

# Let's start by plotting our time series.
plt.title(prog_1)
for id_equipment in range(15):
    mask_equip1 = df['Equipment'] == id_equipment# Select column Equipment with value x
    signal = df.loc[mask_equip1,'2']
    plt.plot(range(len(signal)), signal )
plt.show()

plt.title(prog_2)
# Let's start by plotting our time series.
for id_equipment in range(15):
    mask_equip1 = df['Equipment'] == id_equipment# Select column Equipment with value x
    signal = df.loc[mask_equip1,'3']
    plt.plot(range(len(signal)), signal )
plt.show()
```
